```python
from datetime import datetime
```

```python
data = pd.read_csv('pydata-book/examples/spx.csv',index_col=0,parse_dates=True)
```

```python
data.head()
```

```python
spx = data['SPX']
```

```python
spx[:3]
```

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2,1,1)
ax1.plot(spx)
ax1.set_xlim(['1/1/2007','1/1/2011'])
ax1 = fig.add_subplot(2,1,2)
spx.plot()
ax1.set_xlim(['1/1/2007','1/1/2011'])
```

```python
fig = plt.figure(figsize=(18,9))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(spx)
ax1.set_xlim(['1/1/2007','1/1/2011'])
cirsis_data = [(datetime(2007,10,11),'peak of bull market'),
               (datetime(2008,3,12),'bear stearns fails'),(datetime(2008,9,15),'lehman bankruptcy')]
for date,label in cirsis_data:
    print(spx.asof(date))
    ax1.annotate(label,xy=(date,spx.asof(date)+50),xytext=(date,spx.asof(date)+200),arrowprops=dict(facecolor='black'))
ax1.set_title('important dates in 2008-2009 financial crisis')
```
